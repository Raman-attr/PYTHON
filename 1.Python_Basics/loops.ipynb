Conditional Statement and Loops Question Practice

```python
## Ques 1 = Write a program that ask the user to input the number and prints whether the number is positive

user_number = float(input("Enter the number"))
if user_number>0:
    print("The number is positive")
```

```python
##Ques 2 = Write a program that ask the user to input the number and prints the whether the number is positive or negative

number = float(input("Enter the number"))
if number>0 : 
    print("The number is positive")
elif number<0:
    print("The number is negative")
else:
    print("The number is zero")        
```

```python
## Ques 3 - Write the program that ask the user to input the number and print the whether the number is positive and even , positive and odd , and negative

number = float(input("Enter the number"))

if number>0:
    if (number%2==0):
        print("The number is positive and even")
    else:
        print("The  number is positive and odd")
else:
    print("The number is negative")            
```

```python
## LOOPS

## Ques 1 - Write a program that print all the number from 1 to 10 using a for loop

for i in range(1,11):
    print(i)
```

```python
## Ques 2 - Write a program that print all the number from 1 to 10 using while loop
i = 1
while i<=10:
    print(i)
    i+=1
```

```python
## Ques 3 - Write a  program that prints all the number from 1 to 10 except 5 using a for loop and continue statement
for i in range (1,11):
    if i == 5:
        continue
    print (i)
```
