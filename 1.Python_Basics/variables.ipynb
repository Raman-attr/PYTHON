Variables

Variables are fundamental elements in programming used to store data that can be reference and manipulated in a program . In python , variables are created when you assign a value to them  and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
## Declaring and assigning variables
age = 22
name = "Raman"
height = 5.7
is_student = True

## printing the variables
print("age :" , age )
print("height :" , height)
print("Name : " , name)
```

```python
## Ques 4 - Write a Python Program to find the largest three number
num1 = float(input("Enter the first number"))
num2 =  float(input("Enter the second number"))
num3 =  float(input("Enter the third number"))

if (num1>= num2) and (num1>=num3):
    largest = num1
elif (num2>=num1) and (num2>=num1):
    largest = num2
else:
    largest = num3

print(f"The largest number is : {largest}")

```
