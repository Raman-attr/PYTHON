Pandas - DataFrame and Series

Pandas is a powerful data manipulation library in python , widely used for data anaylsis and data cleaning. It provides two primary data structures . A series is a one-dimensional array -like object , while a DataFrame is a two-dimensional  , size-mutable , and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
```

```python
## Series
## Pandas Series is a  one-dimensional array - like object that can hold any data type . It is similar to a column in a table.

data = [1,2,3,4,5]
series = pd.Series(data)
print('series \n',series)
```

```python
##  Create  a Series from dictionary
data = {'a':1,'b':2,'c':3}
series_dict =pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index = [1,2,3]
pd.Series(data , index = index)
```

```python
# DataFrame
## Create a DataFrame from a dictionary of list

data = {
    'Name':['Raman','Sachin','Jack'],
    'Age' : [22,24,45],
    'City' : ['Delhi','New York','Florida']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a Data frame from a list of Dictionaries

data = [
    {'Name':'Raman','Age':22,'City':'Delhi'},
    {'Name':'Jack','Age':32,'City':'Delhi'},
    {'Name':'Bappy','Age':42,'City':'Delhi'},
    {'Name':'Tom','Age':52,'City':'Delhi'},
]
df = pd.DataFrame(data)
print(df)
print(type(df))
```
